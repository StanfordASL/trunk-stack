```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
import csv

with open('times.csv', 'r') as file:
    reader = csv.reader(file)
    data = next(reader)  # Read first (and only) row
    times = [float(x) for x in data]  # Convert strings to floats

import csv

with open('values.csv', 'r') as file:
    reader = csv.reader(file)
    values = [[float(x) for x in row] for row in reader]

times = np.array(times)
values = np.array(values)
```

```python
plt.plot(times, values[:,0], label='Measured input')
plt.axhline(y=30, color='r', linestyle='--', label="Input reference")
plt.axvline(x=0, color='k', linestyle='-', alpha=0.2)
plt.xlabel("Delay [s]")
plt.ylabel("Input value")
plt.xlim(-0.1, 0.5)
plt.title("Upper bound on delay")
plt.legend()
```
